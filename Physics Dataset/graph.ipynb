```python
data = physics_data_csv
x = data["Distance from the charge to the Equipotential(m)"]
y = data["Magnitude of the Electric Field(V/m)"]
z = data["Voltage of Equipotential surfaces(V)"]
```

```python
x = data["Distance from the charge to the Equipotential(m)"]
y = data["Magnitude of the Electric Field(V/m)"]
z = data["Voltage of Equipotential surfaces(V)"]
w = 7
h = 5
d = 150
fig = plt.figure(figsize=(w, h), dpi=d)
plt.xlim(0, 4.5)
plt.plot(x,z)
fig.suptitle('Distance VS Voltage')
plt.xlabel('Distance from the charge to the Equipotential(m)')
plt.ylabel('Voltage of Equipotential surfaces(V)')
plt.xticks(rotation=45)
ax = plt.gca()
f = lambda x,pos: str(x).rstrip('0').rstrip('.')
ax.xaxis.set_major_locator(matplotlib.ticker.MultipleLocator(0.25))
ax.xaxis.set_major_formatter(matplotlib.ticker.FuncFormatter(f))
l = lambda y,pos: str(y).rstrip('0').rstrip('.')
ax.yaxis.set_major_locator(matplotlib.ticker.MultipleLocator(2))
ax.yaxis.set_major_formatter(matplotlib.ticker.FuncFormatter(f))
plt.savefig("out.png")
plt.show()

```
